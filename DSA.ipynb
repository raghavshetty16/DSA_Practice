2.	Write a program to perform matrix multiplication of two arrays.

```python
def multiply_matrices(arr1, arr2):
    rows1=len(arr1)
    cols1=len(arr1[0])
    rows2=len(arr2)
    cols2=len(arr2[0])
    if cols1!=rows2:
        print("Cannot multiply the arrays")
        return None
    result=[[0 for i in range(cols2)] for j in range(rows1)]
    for i in range(rows1):
        for j in range(cols2):
            for k in range(cols1):
                result[i][j]+=arr1[i][k]*arr2[k][j]
    return result

arr1 = [[1, 2, 3], [4, 5, 6]]
arr2 = [[7, 8], [9, 10], [11, 12]]

result = multiply_matrices(arr1, arr2)
if result:
    print("Result matrix:")
    for row in result:
        print(row)
```

3.	Write a program to convert a matrix into its sparse representation.

```python
def sparse(matrix):
    rows, cols = len(matrix), len(matrix[0])
    sparse = []
    count = 0

    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] != 0:
                count += 1

    sparse.append([rows, cols, count])

    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] != 0:
                sparse.append([i, j, matrix[i][j]])

    return sparse
```

```python
matrix = [
    [0, 0, 3],
    [4, 0, 0],
    [0, 5, 0]
]
res = sparse(matrix)
res
```

```python
def sparse1(matrix):
    rows, cols = len(matrix), len(matrix[0])
    sparse = []
    
    count = 0
    data = []

    for i in range(rows):
        for j in range(cols):
            if matrix[i][j] != 0:
                data.append([i, j, matrix[i][j]])
                count += 1

    sparse.append([rows, cols, count])
    sparse.extend(data)

    return sparse
```

4.	Write a program to find the transpose of a sparse matrix.

```python
def transpose(sparse):
    rows, cols, count = sparse[0]

    result = [[cols, rows, count]] + [[0, 0, 0] for _ in range(count)]

    col_count = [0] * cols
    for r, c, _ in sparse[1:]:
        col_count[c] += 1

    position = [0] * cols
    position[0] = 1

    for i in range(1, cols):
        position[i] = position[i-1] + col_count[i-1]

    for r, c, val in sparse[1:]:
        index = position[c]
        result[index] = [c, r, val]
        position[c] += 1

    return result
```

```python
sparse = [
    [3, 3, 3],
    [0, 2, 3],
    [1, 0, 4],
    [2, 1, 5]
]
res = transpose(sparse)
res
```

```python
n = int(input("Enter number of terms: "))

poly = []

for i in range(n):
    coeff = int(input("Enter coefficient: "))
    power = int(input("Enter power: "))

    poly.append((coeff, power))

terms = []

for coeff, power in poly:

    if power == 0:
        terms.append(f"{coeff}")

    elif power == 1:
        terms.append(f"{coeff}x")

    else:
        terms.append(f"{coeff}x^{power}")

print("\nPolynomial is:")
print(" + ".join(terms))
```

6.	Write a program to store multiple polynomial expressions with different exponents.

```python
num_poly = int(input("Enter number of polynomials: "))

polynomials = []

for i in range(num_poly):

    n = int(input("Enter number of terms: "))

    poly = []

    for j in range(n):

        coeff = int(input("Enter coefficient: "))
        exp = int(input("Enter exponent: "))

        poly.append([coeff, exp])

    polynomials.append(poly)

print("\nStored Polynomials:\n")

for i, poly in enumerate(polynomials, start=1):

    print(f"P{i} =", end=" ")

    terms = []

    for coeff, exp in poly:

        if exp == 0:
            terms.append(f"{coeff}")

        elif exp == 1:
            terms.append(f"{coeff}x")

        else:
            terms.append(f"{coeff}x^{exp}")

    print(" + ".join(terms))
```

```python
def findMedianSortedArrays(A, B):
    if len(A) > len(B):
        return findMedianSortedArrays(B, A)  # ensure A is smaller

    n, m = len(A), len(B)
    lo, hi = 0, n

    while lo <= hi:
        i = (lo + hi) // 2          # partition index in A
        j = (n + m + 1) // 2 - i   # partition index in B

        maxLeftA  = float('-inf') if i == 0 else A[i-1]
        minRightA = float('inf')  if i == n else A[i]
        maxLeftB  = float('-inf') if j == 0 else B[j-1]
        minRightB = float('inf')  if j == m else B[j]

        if maxLeftA <= minRightB and maxLeftB <= minRightA:
            if (n + m) % 2 == 1:
                return max(maxLeftA, maxLeftB)
            else:
                return (max(maxLeftA, maxLeftB) + min(minRightA, minRightB)) / 2
        elif maxLeftA > minRightB:
            hi = i - 1
        else:
            lo = i + 1
```

```python
A = [1, 3, 8]
B = [2, 4, 6, 7]
print(findMedianSortedArrays(A, B))
```

```python
def maxMinPrint(arr):
    arr.sort()
    result = []
    lo, hi = 0, len(arr) - 1
    turn = True  # True = pick max, False = pick min
    while lo <= hi:
        if turn:
            result.append(arr[hi])
            hi -= 1
        else:
            result.append(arr[lo])
            lo += 1
        turn = not turn
    print(result)
```

```python
maxMinPrint([1, 2, 3, 4, 5])
maxMinPrint([10, 30, 20, 50, 40])
maxMinPrint([3, 1, 4, 1, 5, 9]) 
```

```python
def copy_stack(S):
    temp = []
    T = []

    # Empty S into temp (reverses order)
    while S:
        temp.append(S.pop())

    # Empty temp into both T and restore S
    while temp:
        val = temp.pop()
        S.append(val)
        T.append(val)

    return T
```

```python
S = [1, 2, 3]  
T = copy_stack(S)
print("S:", S)  
print("T:", T)
```

```python
class BrowserHistory:
    def __init__(self, homepage):
        self.current = homepage
        self.back_stack = []
        self.forward_stack = []

    def visit(self, url):
        self.back_stack.append(self.current)
        self.current = url
        self.forward_stack = []

    def back(self, steps):
        while steps > 0 and self.back_stack:
            self.forward_stack.append(self.current)
            self.current = self.back_stack.pop()
            steps -= 1
        return self.current

    def forward(self, steps):
        while steps > 0 and self.forward_stack:
            self.back_stack.append(self.current)
            self.current = self.forward_stack.pop()
            steps -= 1
        return self.current  
```

```python
b = BrowserHistory("homepage.com")

b.visit("google.com")
b.visit("youtube.com")
b.visit("github.com")

print(b.back(1))     
print(b.back(1))     
print(b.forward(1))  
b.visit("reddit.com")
print(b.forward(1))  
print(b.back(2))     
```
