A.	Write a program for implementing the following sorting techniques to arrange a list of integers in ascending order. Also demonstrate a comparison table (Data movement and Data Comparison) of different Sorting Algorithms

a.	Quick sort

```python
# Function to find the partition position
def partition(array, low, high):
    # Choose the rightmost element as pivot
    pivot = array[high]

    # Pointer for smaller element
    i = low - 1

    # Traverse through all elements
    # Compare each element with pivot
    for j in range(low, high):
        if array[j] <= pivot:
            # If element is smaller than or equal to pivot,
            # increment i and swap array[i] with array[j]
            i += 1
            array[i], array[j] = array[j], array[i]

    # Place pivot in the correct position
    array[i + 1], array[high] = array[high], array[i + 1]

    # Return pivot index
    return i + 1


# Function to perform Quick Sort
def quicksort(array, low, high):
    if low < high:
        # Partition the array and get pivot index
        pi = partition(array, low, high)

        # Recursively sort elements before pivot
        quicksort(array, low, pi - 1)

        # Recursively sort elements after pivot
        quicksort(array, pi + 1, high)


# Driver Code
if __name__ == "__main__":
    array = [9, 1, 2, 3, 4, 5]

    # Call Quick Sort
    quicksort(array, 0, len(array) - 1)

    # Print sorted array
    print("Sorted Array:", array)
```

b.	Merge sort

```python
# Merge Sort Function
def merge_sort(array):
    # Divide the array until each subarray contains one element
    if len(array) > 1:
        mid = len(array) // 2

        # Split the array into two halves
        left = array[:mid]
        right = array[mid:]

        # Recursively sort both halves
        merge_sort(left)
        merge_sort(right)

        # Initialize pointers
        i = 0  # Pointer for left subarray
        j = 0  # Pointer for right subarray
        k = 0  # Pointer for main array

        # Merge the two sorted subarrays
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                array[k] = left[i]
                i += 1
            else:
                array[k] = right[j]
                j += 1
            k += 1

        # Copy any remaining elements from left subarray
        while i < len(left):
            array[k] = left[i]
            i += 1
            k += 1

        # Copy any remaining elements from right subarray
        while j < len(right):
            array[k] = right[j]
            j += 1
            k += 1


# Driver Code
if __name__ == "__main__":
    array = [10, 1, 9, 7, 5, 4, 3, 2, 6]

    print("Original Array:", array)

    # Call Merge Sort
    merge_sort(array)

    print("Sorted Array:", array)
```

B.	Find median of an array using quicksort. You cannot sort the entire array. Additionally, if it is evident that median is in a particular half of the array, you must stop sorting the other half of the array.

```python
import random


def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1


def quick_select(arr, low, high, k):
    if low <= high:
        pivot_index = partition(arr, low, high)

        if pivot_index == k:
            return arr[pivot_index]
        elif k < pivot_index:
            return quick_select(arr, low, pivot_index - 1, k)
        else:
            return quick_select(arr, pivot_index + 1, high, k)


def find_median(arr):
    n = len(arr)

    if n % 2 == 1:
        return quick_select(arr, 0, n - 1, n // 2)
    else:
        left_middle = quick_select(arr[:], 0, n - 1, n // 2 - 1)
        right_middle = quick_select(arr[:], 0, n - 1, n // 2)
        return (left_middle + right_middle) / 2


if __name__ == "__main__":
    n = int(input("Enter the size of the array: "))

    array = [random.randint(1, 10000) for _ in range(n)]

    print("Original Array:", array)

    median = find_median(array[:])

    print("Median:", median)
```

F.	Implement stack using queue and queue using stack.

```python
import random


# ---------------- STACK USING QUEUE ----------------
class StackUsingQueue:
    def __init__(self):
        self.queue = []

    def push(self, value):
        self.queue.append(value)

    def pop(self):
        if len(self.queue) == 0:
            print("Stack Underflow")
            return

        for _ in range(len(self.queue) - 1):
            self.queue.append(self.queue.pop(0))

        print("Popped element:", self.queue.pop(0))

    def top(self):
        if len(self.queue) == 0:
            print("Stack is empty")
        else:
            print("Top element:", self.queue[-1])

    def display(self):
        print("Stack:", self.queue)


# ---------------- QUEUE USING STACK ----------------
class QueueUsingStack:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, value):
        self.stack1.append(value)

    def dequeue(self):
        if len(self.stack1) == 0 and len(self.stack2) == 0:
            print("Queue Underflow")
            return

        if len(self.stack2) == 0:
            while len(self.stack1) > 0:
                self.stack2.append(self.stack1.pop())

        print("Dequeued element:", self.stack2.pop())

    def front(self):
        if len(self.stack1) == 0 and len(self.stack2) == 0:
            print("Queue is empty")
            return

        if len(self.stack2) == 0:
            while len(self.stack1) > 0:
                self.stack2.append(self.stack1.pop())

        print("Front element:", self.stack2[-1])

    def display(self):
        queue = self.stack2[::-1] + self.stack1
        print("Queue:", queue)


# ---------------- DRIVER CODE ----------------
if __name__ == "__main__":
    n = int(input("Enter the number of random elements: "))

    random_values = [random.randint(1, 100) for _ in range(n)]
    print("Random Values:", random_values)

    print("\n--- Stack Using Queue ---")
    stack = StackUsingQueue()

    for value in random_values:
        stack.push(value)

    stack.display()
    stack.top()
    stack.pop()
    stack.display()

    print("\n--- Queue Using Stack ---")
    queue = QueueUsingStack()

    for value in random_values:
        queue.enqueue(value)

    queue.display()
    queue.front()
    queue.dequeue()
    queue.display()
```

A.	Implementation of the “Tower of Hanoi (Recursive and non-Recursive)” Algorithms to create a program in Python. And Print tower of Hanoi steps.

Recursive Tower of Hanoi

```python
import random


def tower_of_hanoi(n, source, auxiliary, destination):
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return

    tower_of_hanoi(n - 1, source, destination, auxiliary)
    print(f"Move disk {n} from {source} to {destination}")
    tower_of_hanoi(n - 1, auxiliary, source, destination)


if __name__ == "__main__":
    n = random.randint(2, 5)

    print("Number of disks (randomly generated):", n)
    print("\nRecursive Tower of Hanoi Steps:\n")

    tower_of_hanoi(n, "A", "B", "C")
```

Non-Recursive Tower of Hanoi

```python
import random


def move_between_poles(src, dest, src_name, dest_name):
    if len(src) == 0:
        disk = dest.pop()
        src.append(disk)
        print(f"Move disk {disk} from {dest_name} to {src_name}")

    elif len(dest) == 0:
        disk = src.pop()
        dest.append(disk)
        print(f"Move disk {disk} from {src_name} to {dest_name}")

    elif src[-1] < dest[-1]:
        disk = src.pop()
        dest.append(disk)
        print(f"Move disk {disk} from {src_name} to {dest_name}")

    else:
        disk = dest.pop()
        src.append(disk)
        print(f"Move disk {disk} from {dest_name} to {src_name}")


def tower_of_hanoi_non_recursive(n):
    source = list(range(n, 0, -1))
    auxiliary = []
    destination = []

    source_name = "A"
    auxiliary_name = "B"
    destination_name = "C"

    if n % 2 == 0:
        auxiliary_name, destination_name = destination_name, auxiliary_name
        auxiliary, destination = destination, auxiliary

    total_moves = 2 ** n - 1

    for move in range(1, total_moves + 1):
        if move % 3 == 1:
            move_between_poles(source, destination, source_name, destination_name)
        elif move % 3 == 2:
            move_between_poles(source, auxiliary, source_name, auxiliary_name)
        else:
            move_between_poles(auxiliary, destination, auxiliary_name, destination_name)


if __name__ == "__main__":
    n = random.randint(2, 5)

    print("Number of disks (randomly generated):", n)
    print("\nNon-Recursive Tower of Hanoi Steps:\n")

    tower_of_hanoi_non_recursive(n)
```

```python
import random


def precedence(operator):
    if operator in ['+', '-']:
        return 1
    elif operator in ['*', '/']:
        return 2
    elif operator == '^':
        return 3
    return 0


def infix_to_postfix(expression):
    stack = []
    postfix = ""

    for ch in expression:
        if ch.isalnum():
            postfix += ch

        elif ch == '(':
            stack.append(ch)

        elif ch == ')':
            while stack and stack[-1] != '(':
                postfix += stack.pop()
            if stack:
                stack.pop()

        else:
            while stack and precedence(stack[-1]) >= precedence(ch):
                postfix += stack.pop()
            stack.append(ch)

    while stack:
        postfix += stack.pop()

    return postfix


def generate_random_infix():
    operands = ['A', 'B', 'C', 'D', 'E', 'F']
    operators = ['+', '-', '*', '/']

    num_operands = random.randint(3, 5)
    selected_operands = random.sample(operands, num_operands)

    expression = selected_operands[0]

    for i in range(1, num_operands):
        op = random.choice(operators)
        expression += op + selected_operands[i]

    return expression


if __name__ == "__main__":
    infix = generate_random_infix()

    print("Random Infix Expression :", infix)

    postfix = infix_to_postfix(infix)

    print("Postfix Expression      :", postfix)
```

C.	Implementation of the “Postfix Evaluation” Algorithm to create a program in Python.

```python
import random


def evaluate_postfix(expression):
    stack = []

    for ch in expression:
        if ch.isdigit():
            stack.append(int(ch))
        else:
            b = stack.pop()
            a = stack.pop()

            if ch == '+':
                stack.append(a + b)
            elif ch == '-':
                stack.append(a - b)
            elif ch == '*':
                stack.append(a * b)
            elif ch == '/':
                stack.append(a / b)
            elif ch == '^':
                stack.append(a ** b)

    return stack.pop()


def generate_random_postfix():
    a = random.randint(1, 9)
    b = random.randint(1, 9)
    c = random.randint(1, 9)

    operators = ['+', '-', '*', '/']
    op1 = random.choice(operators)
    op2 = random.choice(operators)

    postfix = f"{a}{b}{op1}{c}{op2}"
    return postfix


if __name__ == "__main__":
    postfix = generate_random_postfix()

    print("Random Postfix Expression :", postfix)

    result = evaluate_postfix(postfix)

    print("Evaluation Result         :", result)
```

```python
import random


def evaluate_postfix(expression, values):
    stack = []

    for ch in expression:
        if ch.isalpha():
            stack.append(values[ch])
        else:
            b = stack.pop()
            a = stack.pop()

            if ch == '+':
                stack.append(a + b)
            elif ch == '-':
                stack.append(a - b)
            elif ch == '*':
                stack.append(a * b)
            elif ch == '/':
                stack.append(a / b)
            elif ch == '^':
                stack.append(a ** b)

    return stack.pop()


def generate_random_postfix():
    operands = ['A', 'B', 'C', 'D', 'E']
    operators = ['+', '-', '*', '/']

    selected_operands = random.sample(operands, 3)

    op1 = random.choice(operators)
    op2 = random.choice(operators)

    postfix = (
        selected_operands[0]
        + selected_operands[1]
        + op1
        + selected_operands[2]
        + op2
    )

    return postfix, selected_operands


if __name__ == "__main__":
    postfix, operands = generate_random_postfix()

    print("Random Postfix Expression :", postfix)

    values = {}
    for operand in operands:
        values[operand] = random.randint(1, 9)

    print("Assigned Values:")
    for operand in operands:
        print(operand, "=", values[operand])

    result = evaluate_postfix(postfix, values)

    print("Evaluation Result         :", result)
```
