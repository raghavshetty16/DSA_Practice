```python
import random


def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key


def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1


# Generate random array size
n = random.randint(5, 10000)

# Generate random array elements
arr = [random.randint(-1000000, 1000000) for _ in range(n)]

print("Randomly Generated Array:")
print(arr)

# Sort the array
insertion_sort(arr)

print("\nSorted Array:")
print(arr)

# Select a target that is guaranteed to be present in the array
target = random.choice(arr)

print("\nElement to search:", target)

# Perform binary search
index = binary_search(arr, target)

if index != -1:
    print(f"Element {target} found at index {index}.")
else:
    print(f"Element {target} not found in the array.")
```
