B. Implementation of the “Insertion, Selection Sort & Binary Search” Algorithm to create a program in C/python

```python
array = [191, 181, 171, 161, 151, 141, 131, 121, 111, 101]

print(array)

for i in range(1, len(array)):
    key = array[i]
    j = i - 1

    while j >= 0 and key < array[j]:
        array[j + 1] = array[j]
        array[j] = key
        key = array[j]
        j = j - 1

print(array)
```

C. Take a randomly generated array and sort it using insertion sort. Then find an element using binary search.

```python
import random

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

n = int(input("Enter the size of the array: "))

arr = [random.randint(1, 100) for _ in range(n)]

print("Randomly Generated Array:")
print(arr)

insertion_sort(arr)

print("Sorted Array (using Insertion Sort):")
print(arr)

target = int(input("Enter the element to search: "))

index = binary_search(arr, target)

if index != -1:
    print(f"Element {target} found at index {index}.")
else:
    print(f"Element {target} not found in the array.")
```

D. Take a randomly generated array and sort it using selection sort. Then find an element using binary search. 

```python
import random

def selection_sort(arr):
    for i in range(len(arr)):
        min_index = i

        for j in range(i + 1, len(arr)):
            if arr[min_index] > arr[j]:
                min_index = j

        arr[i], arr[min_index] = arr[min_index], arr[i]

def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

n = int(input("Enter the size of the array: "))

arr = [random.randint(1, 100) for _ in range(n)]

print("Randomly Generated Array:")
print(arr)

selection_sort(arr)

print("Sorted Array (using Selection Sort):")
print(arr)

target = int(input("Enter the element to search: "))

index = binary_search(arr, target)

if index != -1:
    print(f"Element {target} found at index {index}.")
else:
    print(f"Element {target} not found in the array.")
```

E. Take a randomly generated array and reverse-sort it using insertion sort. Then find an element using binary search. 

```python
import random

def insertion_sort_desc(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] < key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

def binary_search_desc(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

n = int(input("Enter the size of the array: "))

arr = [random.randint(1, 100) for _ in range(n)]

print("Randomly Generated Array:")
print(arr)

insertion_sort_desc(arr)

print("Reverse Sorted Array (using Insertion Sort):")
print(arr)

target = int(input("Enter the element to search: "))

index = binary_search_desc(arr, target)

if index != -1:
    print(f"Element {target} found at index {index}.")
else:
    print(f"Element {target} not found in the array.")
```

F. Take a randomly generated array and reverse-sort it is using bubble sort. Then find an element using linear search. 

```python
import random

def bubble_sort_desc(arr):
    for i in range(len(arr)):
        for j in range(len(arr) - i - 1):
            if arr[j] < arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

n = int(input("Enter the size of the array: "))

arr = [random.randint(1, 100) for _ in range(n)]

print("Randomly Generated Array:")
print(arr)

bubble_sort_desc(arr)

print("Reverse Sorted Array (using Bubble Sort):")
print(arr)

target = int(input("Enter the element to search: "))

index = linear_search(arr, target)

if index != -1:
    print(f"Element {target} found at index {index}.")
else:
    print(f"Element {target} not found in the array.")
```

G. Take a randomly generated array and sort the 2nd half using bubble sort. Then find the median using linear search. 

```python
import random

def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(len(arr) - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

n = int(input("Enter the size of the array: "))

arr = [random.randint(1, 100) for _ in range(n)]

print("Randomly Generated Array:")
print(arr)

mid = len(arr) // 2

second_half = arr[mid:]
bubble_sort(second_half)
arr[mid:] = second_half

print("Array after sorting the 2nd half using Bubble Sort:")
print(arr)

sorted_arr = arr[:]
bubble_sort(sorted_arr)

if len(sorted_arr) % 2 == 0:
    median = (sorted_arr[len(sorted_arr)//2 - 1] + sorted_arr[len(sorted_arr)//2]) / 2
else:
    median = sorted_arr[len(sorted_arr)//2]

print("Median value:", median)

index = linear_search(arr, median)

if index != -1:
    print(f"Median found at index {index}.")
else:
    print("Median is not present as an exact element in the array.")
```
